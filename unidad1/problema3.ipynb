Imagina que estás organizando un concurso de dibujo para niños en una feria local. Quieres diseñar un programa que genere automáticamente patrones para colorear. Cada patrón debe consistir en una serie de figuras geométricas simples, como círculos, cuadrados y triángulos, dispuestos de manera creativa. Además, el programa debe permitir ajustar el número de figuras y su tamaño para crear una variedad de patrones únicos. ¿Cómo diseñarías un algoritmo que genere estos patrones y permita ajustar los parámetros para crear diseños diferentes y atractivos?

```
Inicio
   |
   |---> Solicitar: número de figuras, tamaño, tipo de figuras
   |
   |---> Generar patrón con las figuras especificadas
   |
   |---> Ajustar tamaño y número de figuras
   |
   |---> Mostrar patrón generado
   |
Fin
```

````python
import matplotlib.pyplot as plt
import numpy as np
import random
````

````python
# Función para generar un patrón de figuras geométricas

def generar_patron(num_figuras=10, tamaño=5):
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.set_xticks([])
    ax.set_yticks([])
    
    for _ in range(num_figuras):
        x, y = random.uniform(1, 9), random.uniform(1, 9)
        tipo = random.choice(["circulo", "cuadrado", "triángulo"])
        
        if tipo == "circulo":
            circle = plt.Circle((x, y), tamaño / 10, color=random.choice(["red", "blue", "green"]), fill=True)
            ax.add_patch(circle)
        elif tipo == "cuadrado":
            square = plt.Rectangle((x - tamaño / 20, y - tamaño / 20), tamaño / 10, tamaño / 10, color=random.choice(["purple", "orange", "cyan"]), fill=True)
            ax.add_patch(square)
        elif tipo == "triángulo":
            triangle = plt.Polygon([(x, y + tamaño / 10), (x - tamaño / 20, y - tamaño / 20), (x + tamaño / 20, y - tamaño / 20)], color=random.choice(["pink", "yellow", "gray"]), fill=True)
            ax.add_patch(triangle)

    plt.title("Patrón de figuras geométricas")
    plt.show()

# Generar un patrón con 15 figuras de tamaño 7
generar_patron(15, 7)
````

````python
# Generar un patrón para colorear

def generar_patron_contorno(num_figuras=10, tamaño=5):
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.set_xticks([])
    ax.set_yticks([])

    for _ in range(num_figuras):
        x, y = random.uniform(1, 9), random.uniform(1, 9)
        tipo = random.choice(["circulo", "cuadrado", "triángulo"])

        if tipo == "circulo":
            circle = plt.Circle((x, y), tamaño / 10, edgecolor='black', fill=False)
            ax.add_patch(circle)
        elif tipo == "cuadrado":
            square = plt.Rectangle((x - tamaño / 20, y - tamaño / 20), tamaño / 10, tamaño / 10, edgecolor='black', fill=False)
            ax.add_patch(square)
        elif tipo == "triángulo":
            triangle = plt.Polygon([(x, y + tamaño / 10), (x - tamaño / 20, y - tamaño / 20), (x + tamaño / 20, y - tamaño / 20)], edgecolor='black', fill=False)
            ax.add_patch(triangle)

    plt.title("Patrón para colorear")
    plt.show()

# Generar un patrón con 15 figuras de tamaño 7 para colorear
generar_patron_contorno(15, 7)
````
